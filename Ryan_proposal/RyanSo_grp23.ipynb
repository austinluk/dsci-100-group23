2. Question:


Our group is interested in identifying what player characteristics and behaviors are most predictive of subscribing to a game-related newsletter, and how do these feature differ between various player types. 



In order to find out what player characteristics and behaviours are most predictive of subscribing to a game-related newsletter, we can filter out the age group and playing hours of the players. 
In this way we can find if there is correlation between age group and playing hours.We can also use a histogram to see the distribution of experience and subscription to see if incentive to subscribe is related to level
of experience. But before any of these can be done, we have to first tidy our datasets. 

```python
import altair as alt
import numpy as np
import pandas as pd

#3: Exploratory Data Analysis and Visualization
```

```python
sessions = pd.read_csv("dsci-100-student-python/materials/Grp23_project/data/sessions.csv")
sessions
```

```python
players = pd.read_csv('dsci-100-student-python/materials/Grp23_project/data/players.csv')
players
```

```python
tidy_sessions = sessions.groupby(['start_time', 'end_time']).mean(numeric_only = True).reset_index()
tidy_sessions
```

We can later use this tidied dataset to analyze the time used for each session. Since the date and time are in order, it makes it easier for us to use afterwards. 

We can also make some plots out of the players dataset at this point:

```python
#players_exp = players.loc[players['subscribe']==True,['experience', 'subscribe']]
players_exp = players[['experience','played_hours']]
players_exp
```

```python
exp_plot = alt.Chart(players_exp).mark_point(opacity=0.4).encode(
    x='experience',
    y='played_hours',
    color='experience'
)
exp_plot
```

From the above plot we can see that a lot of players played less than 20 hours. A regular player has the most playing hours at above 220 hours. 

```python
players_sub = players[['played_hours','subscribe']]
players_sub
```

```python
sub_plot = alt.Chart(players_sub).mark_bar(opacity=0.4).encode(
    x='played_hours:N',
    y='subscribe',
    color='subscribe'
)
sub_plot
```

From the above bar plot we can see more subscribers than non-subscribers. A big portion of players had 0-0.1 hours of playing. 

```python
players_age = players[['played_hours','age']]
players_age
```

```python
age_plot = alt.Chart(players_age).mark_circle(opacity=0.4).encode(
    x='age',
    y='played_hours'
)
age_plot
```

From the above scatter plot we can see that most of the players were between 15-26 years old, there are also players who are over 90 years old.

Predictions will be made by utilizing the .predict function in scikit-learn. However, a limitation in the method that may affect our results is its sensitivity to outliers. In players.csv, we can see that there are players that are over 90 years old. This may affect our regression line as their age is significantly older than the majority of the players. Therefore, in order to obtain the most accurate results, we can compare our model to another method which is the knn-regression. In this way we can cross check the results from multiple methods to avoid confirmation bias. Before performing linear regression.
